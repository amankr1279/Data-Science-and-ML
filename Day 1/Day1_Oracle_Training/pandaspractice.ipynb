```python
np.random.seed(2)
pp = pers.sample(frac = 1).reset_index(drop = True)
pp.head()
```

#### Inner Merge --> only keeps the rows for matching rowsfor Primary key

```python
pd.merge(p1,m1, on = 'Enrol No.', how = 'inner')
```

### Outer Merge : all the matching as well unmatching rows are included

```python
pd.merge(p1,m1, on = 'Enrol No.', how = 'outer')
```

### left join : includes all the rows of left dataset

```python
p1
```

```python
pd.merge(p1,m1, on = 'Enrol No.', how = 'left')
```

### Right join : All the rows from the right dataset are included

```python
pd.merge(p1,m1, on = 'Enrol No.', how = 'right')
```

```python
m2 = marks2.iloc[[ 8,  4,  9,  19, 1]]
m2
```

```python
pd.merge(p1,m2, on = 'Enrol No.', how = 'right', suffixes= ('_pers','_marks'))
```
